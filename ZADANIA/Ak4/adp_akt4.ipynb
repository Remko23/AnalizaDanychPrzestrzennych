<b>Ćwiczenie 2.</b> Utwórz funkcję o nazwie <i>create_line_geometry</i>, która na podstawie parametru <i>points</i>, będącego listą obiektów typu Point, zwróci obiekt typu LineString. Powinna też weryfikować za pomocą asercji (ang. <i>assert</i>), czy:
- parametr jest listą. Jeżeli nie, to wyświetlić komunikat: `Parametr musi być listą!`,
- lista składa się z co najmniej dwóch elementów. Jeżeli nie, to wyświetlić komunikat: `Linia musi składać się z minimum dwóch punktów!`,
- lista składa się jedynie z obiektów typu Point. Jeżeli nie, to wyświetlić komunikat: `Lista musi składać się z punktów!`.

```python
from shapely.geometry import LineString

def create_line_geometry(points):
    assert isinstance(points, list), "Parametr musi być listą!"
    assert len(points) >= 2, "Linia musi się składać z minimum dwóch punktów"
    for point in points:
        assert point.geom_type == 'Point', "Linia musi się składać z punktów"

    return LineString(points)
```

Przetestuj działanie funkcji.

```python
points = []
points.append(create_point_geometry(0, 0))
points.append(create_point_geometry(1, 1))

line = create_line_geometry(points)
print(line)
print(line.geom_type)
```

<b>Ćwiczenie 3.</b> Utwórz funkcję o nazwie <i>create_polygon_geometry</i>, która na podstawie parametru <i>coordinates</i>, będącego listą krotek współrzędnych, zwróci obiekt typu Polygon. Powinna też weryfikować za pomocą asercji (ang. <i>assert</i>), czy:
- parametr jest listą. Jeżeli nie, to wyświetlić komunikat: `Parametr musi być listą!`,
- lista składa się z co najmniej trzech elementów. Jeżeli nie, to wyświetlić komunikat: `Wielokąt musi składać się z minimum trzech punktów!`,
- lista składa się jedynie z dwuelementowych krotek. Jeżeli nie, to wyświetlić komunikat: `Lista musi składać się z dwuelementowych krotek!`,
- krotki zawierają jedynie wartości typu int albo float. Jeżeli nie, to wyświetlić komunikat: `Krotki muszą zawierać współrzędne!`.

```python
from shapely.geometry import Polygon

def create_polygon_geometry(coordinates):
    assert isinstance(coordinates, list), "Parametr musi być listą!"
    assert len(coordinates) >= 3, "Wielokąt musi się składać z minimum trzech punktów"
    for coordinate in coordinates:
        assert isinstance(coordinate, tuple), "Lista musi składać się z dwuelementowych krotek!"
        assert isinstance(coordinate[0], int or float) and isinstance(coordinate[1], int or float)

    return Polygon(coordinates)
```

Przetestuj działanie funkcji.

```python
coordinates = []
coordinates.append((0, 0))
coordinates.append((1, 1))
coordinates.append((2, 0))

polygon = create_polygon_geometry(coordinates)
print(polygon)
print(polygon.geom_type)
```
