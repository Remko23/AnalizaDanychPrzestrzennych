<b>Ćwiczenie 1.</b> Utwórz interaktywną mapę na podstawie współrzędnych (60.25, 24.8).

```python
import folium

my_map = folium.Map(location=[60.25, 24.8], zoom_start=12)
```

<b>Ćwiczenie 2.</b> Nanieś na interaktywną mapę znacznik o współrzędnych (60.20426, 24.96179).

```python
folium.Marker(location=[60.20426, 24.96179]).add_to(my_map)
my_map
```

<b>Ćwiczenie 3.</b> Ustaw opcję, aby po kliknięciu na znacznik pojawiał się tekst "Kumpula Campus".

```python
folium.Marker(location=[60.20426, 24.96179], popup="Kumpula Campus").add_to(my_map)

my_map
```

<b>Ćwiczenie 4.</b> Zmień znacznik na zieloną ikonę OK.

```python
folium.Marker(
    location=[60.20426, 24.96179],
    popup="Kumpula Campus",
    icon=folium.Icon(color='green', icon='ok')
).add_to(my_map)

my_map
```

<b>Ćwiczenie 5.</b> Wczytaj dane z pliku ``addresses.shp`` do obiektu typu GeoDataFrame o nazwie <i>points</i>.

```python
import geopandas as gpd

points = gpd.read_file("addresses/addresses.shp")
```

<b>Ćwiczenie 6.</b> Nanieś na interaktywną mapę o stylu cartodbpositron punkty ze zmiennej <i>points</i>. Przypisz im nazwę "Public transport stations".

```python
my_map2 = folium.Map(location=[60.25, 24.8], zoom_start=12, tiles='cartodbpositron')

points.explore(
    m=my_map2,
    color='blue',
    style_kwds={'radius': 5},
    name="Public transport stations"
)

my_map2
```

<b>Ćwiczenie 7.</b> Dodaj do mapy możliwość kontroli warstw.

```python
folium.LayerControl().add_to(my_map2)

my_map2
```

<b>Ćwiczenie 8.</b> Wczytaj do zmiennej <i>locations</i> pary współrzędnych (x, y) ze zmiennej <i>points</i>.

```python
locations = list(zip(points.geometry.y, points.geometry.x))
```

<b>Ćwiczenie 9.</b> Nanieś na interaktywną mapę o stylu cartodbpositron zgrupowane znaczniki o współrzędnych ze zmiennej <i>locations</i>.

```python
from folium.plugins import MarkerCluster

my_map3 = folium.Map(location=[60.25, 24.8], zoom_start=12, tiles='cartodbpositron')

marker_cluster = MarkerCluster(locations).add_to(my_map3)

my_map3
```
